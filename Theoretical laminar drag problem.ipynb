```python
import os
import numpy as np
import math as math
from matplotlib import pyplot as plt
```

```python
m = 1.0
g = 9.80

nstep = 10000
timestep = 0.001
ystart =0
vxstart = 0
xstart = 0
vvstart = 50
aystart=9.8
axstart=0
tplot = np.zeros(nstep)
xplot = np.zeros(nstep)
vxplot = np.zeros(nstep)
yplot = np.zeros(nstep)
vvplot = np.zeros(nstep)
ayplot = np.zeros(nstep)
axplot = np.zeros(nstep)
vplot= np.zeros(nstep)
aplot= np.zeros(nstep)
alpha= 0.2
n = 0
ay= aystart
ax= axstart
a= np.sqrt(ax**2 + ay**2)
t = 0
x = xstart
vx= vxstart
vv= vvstart
y=ystart
v=0
tplot[0] = t
xplot[0] = x
yplot[0] = y
vvplot[0] = vv
vxplot[0] = vx
ayplot[0]= ay
axplot[0]= ax
vplot[0]=v
aplot[0]=a

   
for nloop in range(1,nstep):
        n = n+1
        
        t = t + timestep
        y = y + timestep*vv
        vv = vv + timestep*ay/m
        x = x + timestep*vx
        ay= -g-(alpha/m)*vv
        ax= -alpha*vx
        vx= vx + timestep*ax
        v= np.sqrt(vv**2 + vx**2)
        a= np.sqrt(ax**2 + ay**2)
        
        tplot[n] = t
        xplot[n] = x
        vvplot[n] = vv
        yplot[n] = y
        ayplot[n]= ay
        vxplot[n]= vx
        axplot[n]= ax
        vplot[n]=v
        aplot[n]=a
        continue
#
plt.title("Freefall")
plt.xlabel('time')
plt.ylabel('y-position')
plt.scatter(tplot, yplot, s=1)
plt.show
plt.savefig('Position graph',dpi=300)
print(np.max(yplot))
print((np.argmax(yplot))*0.001)
```
